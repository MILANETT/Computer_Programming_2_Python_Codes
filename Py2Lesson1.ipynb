```python
#Lesson 1 INTRO - Using lista
import statistics as stat
x=[60,70,8,90,78]
mn=stat.mean(x)
md=stat.mode(x)
mdn=stat.median(x)
std=stat.stdev(x)
var=stat.variance(x)
print('Summary statistics is as follows')
print(f'The mean is: {mn:.3f}')
print(f'The mode is is: {md:.2f}')
print(f'The median is: {mdn:.2f}')
print(f'The standard deviation is: {std:.2f}')
print(f'The variance is: {var:.2f}')
```

```python
#Lesson 2 NUMPY: using arrays
import numpy as np
x=[50,60,80,80,50]
arr=np.array(x)#using array function to convert list to array
arr_mean=np.mean(arr)
arr_stdev=np.std(arr)
arr_dim=arr.ndim
arr_shape=arr.shape
print(f'The mean is: {arr_mean}')
print(f'The standard deviation is: {arr_stdev}')
print(f"The dimension of the array is: {arr_dim}")
print(f"The shape of the array is: {arr_shape}")
```

```python
#arithmetic operations 24/03/2026
#add
import numpy as np
arr1=[1,3,5,7]
arr2=[2,4,6,8]
first_array=np.array(arr1)
second_array=np.array(arr2)
#using the + operator
result1 = first_array+second_array
print("Using the + operator:", result1)
#using the add() function
result2 = np.add(first_array,second_array)
print("Using the add() function:", result2)
```

```python
#subtract
import numpy as np
arr1=[1,3,5,7]
arr2=[2,4,6,8]
first_array=np.array(arr1)
second_array=np.array(arr2)
#using the - operator
result1 = first_array-second_array
print("Using the - operator:", result1)
#using the subtract function
result2 = np.subtract(first_array,second_array)
print("Using the subtract() function:", result2)
```

```python
#multiply
import numpy as np
arr1=[1,3,5,7]
arr2=[2,4,6,8]
first_array=np.array(arr1)
second_array=np.array(arr2)
#using the * operator
result1 = first_array*second_array
print("Using the * operator:", result1)
#using the multiply() function
result2 = np.multiply(first_array,second_array)
print("Using the multiply() function:", result2)
```

```python
def grading(mean_value_horizontal):
    if mean_value_horizontal>=70 and mean_value_horizontal<=100:
        return 'A'
    
    elif mean_value_horizontal>=60 and mean_value_horizontal<70:
        return'B'
    
    elif mean_value_horizontal>=50 and mean_value_horizontal<60:
        return 'C'
    
    elif mean_value_horizontal>=40 and mean_value_horizontal<50:
        return'D'
    elif mean_value_horizontal>=0 and mean_value_horizontal<40:
        return 'E'
    else:
        return'X'
```

```python
marks_df3['Grade'] = marks_df3['Mean'].apply(grading)
marks_df3
```
